## Kinase object

A de-novo Kinase object get be initiated using a matrix, or simply by calling an existing kinase from the library

```python
kin = kl.get_kinase('AKT1')
```

Kinase object contains information about the retrieved kinase:

```python
print('Name: ' + kin.name)
print('\nMatrix:')
print(kin.log2_matrix)
print('\nKinase type: ' + kin.kin_type)
print('\nKinase family: ' + kin.family)
```

```python
kin.info
```

Plot kinase ID (sequence logo, heatmat, and  S/T favorability):

```python
kin.plot_data()
```

```python
kin.plot_data(title='My Favorite Kinase')
```

Sequence logo can be plotted separately with different configurations

```python
kin.seq_logo(logo_type='prob') #can be 'ratio_to_random', 'ratio_to_median', or 'prob'
```

As well as heatmap

```python
kin.heatmap()
```

Kinase object can be used to quickly score a list of substrates

```python
subs_list = ['VDDEKGDsNDDYDSA',
 'YDSAGLLsDEDCMSV',
 'IADHLFWtEETKSRF',
 'KSRFTEYsMTSSVMR',
 'FTEYSMTsSVMRRNE',
 'TEYSMTSsVMRRNEQ',
 'MCQGGDFtRPNGTGD',
 'MGGRKMAtDEENVYG',
 'ATDEENVsGLEENAQ',
 'FQRGKYGtMAEGRSE']

kin.score(subs_list)
```

```python
kin.percentile(subs_list)
```
