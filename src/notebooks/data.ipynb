### Kinase information

Get the information about an entire kinome:

Get information about the entire kinome:
1. **_MATRIX_NAME_**: name of matrix file
2. **_KINASE_**: kinase name (protein)
3. **_GENE_NAME_**: kinase name (gene)
4. **_TYPE_**: Ser/Thr or Tyrosine
5. **_SUBTYPE_**: for tyrosine kinases - receptor tyrosine kinases (RTK), non receptor tyrosine kinases (nRTK), non-canonical tyrosine kinases (ncTK) [_for Ser/Thr kinases subtype is always STK_]
6. **_FAMILY_**: kinase family
7. **_UNIPROT_ID_**: UniProt ID
8. **_UNIPROT_ENTRY_NAME_**: UniProt entry name
9. **_PDB_ID_**: PBD ID
10. **_KL_LIBRARY_**: type of peptide library used in profiling the kinase<br>
    i. _ser_thr_: excluded amino acids from random mixture - **Ser, Thr, Cys** (total 17 random amino acids, used for Ser/Thr kinases)<br>
    ii. _tyr_: excluded amino acids from random mixture - **Tyr, Cys** (total 18 random amino acids, used for Tyrosine kinases)<br>
    iii. _ser_thr_tyr_: excluded amino acids from random mixture - **Ser, Thr, Tyr, Cys** (total 16 random amino acids, used for kinases with dual specificity)<br>
11. **_DUAL_SPECIFICITY_**: whether the kinase can phosphorylate both Ser/Thr and Tyrosine residues
12. **_DISPLAY_NAME_**: kinase name when displayed on the website

### Scored phosphoproteome information

**In order to calculate the percentile output for a certain kinase, the raw score of the kinase for a given substrate is compared to an _a-priori_ empirical score distribution of an entire phosphoproteome. Different phosphoproteomes can be used (including a customized one), while the Kinase Library default is using the established phosphoproteome described in Ochoa et al. 2020.**

In order to retrieve the list of available phosphoproteomes:

```python
kl.get_phosphoproteomes_list()
```

```python
kl.get_phosphoproteomes_info()
```

The phosphoproteome itself can be retrieved:

```python
kl.get_phosphoproteome('2020-ochoa')
```

And the scored phosphoproteome (with or without info) can be retrieved as well:

```python
kl.get_scored_phosphoproteome(kin_type='ser_thr')
```

```python
kl.get_scored_phosphoproteome(kin_type='tyrosine', with_info=True)
```
